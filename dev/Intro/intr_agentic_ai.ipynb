```python
# Import the library for loading environment variables
from dotenv import load_dotenv

load_dotenv()
```

```python
#Set the override flag to True to allow the environment variables to be overridden
#This is important because only what you have in this environment variable in your folder is given priority.
load_dotenv(override=True)
```

```python
#Import the os library to access the environment variables
import os 
#Get the OpenAI API key from the environment variables
open_api_key = os.getenv("OPENAI_API_KEY")
#Check if the OpenAI API key exists and print the first 8 characters of the key
if open_api_key:
    print(f"OpenAI API key exists and begins {open_api_key[:8]}")
else:
    print("OpenAI API key not set -please fix")    
```

```python
#Import the OpenAI library : a light library that helps you cnnect with openAI in the cloud
from openai import OpenAI

#Create an instance of the OpenAI class
openai = OpenAI()
```

```python
# Create a list of jmesages in the familiar OpenAI formatat
messages = [{"role": "user", "content": "How many countries are in the United Kingdom, please list them"}]

# Call the OpenAI API with the messages
response = openai.chat.completions.create(
    model="gpt-4.1-nano",
    messages=messages
)


# print the response
print(response.choices[0].message.content)
```

```python
question = "Please provide a question assess someone's IQ. Kindly respond only with a question"
messages = [{"role":"user","content":question}]

# Call the OpenAI API wuth messages  and the model 
question = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

#print response as a question
print(question.choices[0].message.content)
```
