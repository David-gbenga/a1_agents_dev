```python
from openai import OpenAI 

#create an instance of openai class
open_ai = OpenAI()
```

```python
question = """What countries exist in the united kingdom, also is 
there election for Prime minister in UK this year and if there is, who are the candidates"""

messages  = [{"role":"user", "content":question}]

request = open_ai.chat.completions.create(

    model = "gpt-4.1-mini",
    messages = messages
)

print(request.choices[0].message.content)
```
